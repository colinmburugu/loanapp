```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('data2.csv')
data.head()
```

```python
corr = data.corr()
corr
```

```python
for i in list(corr.columns):
    for j in list(corr.columns):
        if corr.loc[i,j]>0.7 and corr.loc[i,j] != 1:
            print (i, ' ',j ,' ', corr.loc[i,j]) 
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(columns=['id','def'])
y = data['def']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.35,random_state=0)
```

```python
from sklearn.preprocessing import Imputer
imr = Imputer(missing_values='NaN',strategy= 'mean')
X_train_imr = imr.fit_transform(X_train)
X_test_imr = imr.transform(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_imr,y_train)
```

```python
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance',ascending=False)
feature_importances
```
